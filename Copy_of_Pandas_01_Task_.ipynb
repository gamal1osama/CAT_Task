# **2. CodeForces Problems** 💡



<a href='https://codeforces.com/contest/1883/problem/A'> <font color='blue'>**Problem 1** (Morning)</a> </font>

```python
# Submission link
# https://codeforces.com/contest/1883/submission/249868492

q=int(input())
while(q):
    q-=1
    num=input()
    num=list(num)
    num=list(map(int,num))
    for i in range (4):
        if num[i]==0:
            num[i]=10
    start=num[0]-1
    c=4

    for i in range(1,4):
        c+=abs(num[i]-num[i-1])
    print(start+c)
```

<a href='https://codeforces.com/contest/1878/problem/B'> <font color='blue'>**Problem 2** (Aleksa and Stack)</a> </font>

```python
# Submission link
# https://codeforces.com/contest/1878/submission/249870436
import random
q=int(input())
while(q):
    q-=1
    lis=[1,3]
    num=int(input())
    for i in range(2,num):
        the=random.randint(lis[-1]+1,lis[-1]+10)
        if (the*3)%(lis[i-1]+lis[i-2])==0:
            lis.append(the+1)
        else:
            lis.append(the)
    for i in lis:
        print(i,end=" ")
    print()
```
